```python
import pandas as pd
```

```python
df_salary = pd.read_csv('Employee_Salary_Dataset.csv')
df_department = pd.read_csv('Department_Dataset.csv')
```

```python
df = pd.merge(df_salary, df_department, on='ID')
```

```python
from sklearn.base import BaseEstimator,TransformerMixin
from sklearn.preprocessing import LabelEncoder


class Ordinal(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform (self, X):
        label = LabelEncoder()
        X['Gender'] = label.fit_transform(X['Gender'])
        X['travel_required'] = label.fit_transform(X['travel_required'])
        return X
    
class Dropcolumn(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X = X.drop(['ID', 'location'], axis=1)
        return X

class Categorical(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X = pd.get_dummies(X, columns=['Dept_name'])
        return X
```

```python
## Pipelines
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('ordinal', Ordinal()),
    ('drop_column', Dropcolumn()),
    ('categorical', Categorical())
])
```

```python
df =pipeline.fit_transform(df)
```
